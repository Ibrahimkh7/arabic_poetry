## We realizethat the first element before the ',' is the nature of the poem so let's get the different types present in the dataset

```python
unique = dataset.poem_tags.unique()
len(unique)
```

```python
x = str(unique[0]).split(',')
x[0]
```

```python
types = []
for i in range(len(unique)):
    present = False
    x = str(unique[i]).split(',')
    for j in types:
        if j==x[0]:
            present = True
    if not present and x[0].split(' ')[0] == 'قصائد':
        types.append(x[0])
types
```

```python
keyword_l = []
for i in types:
    name = i.split()[1]
    if name[:2] == 'ال':
        name = name[2:]
    keyword_l.append(name)
keyword_l
```

```python
#the problem in this is that it is al-ta3rif sensible i.e. if they said :عن الغزل this wouldn't work
# We need to filter the words and then remove the ال and then plug it into this
def filtering(s):
    common = ["أعطني", "ابحث", "أريد", "أطلب", "هات", "أرني", "اعثر", "قصيدة", "عن","في","كي"]
    
    wrds = s.split()
    
    filtered = []
    for i in wrds:
        if i not in common:
            filtered.append(i)
    return filtered
```

```python
def get_keyword(s):
    f = filtering(s)
    keyword = ''
    keywords = keyword_l
    
    for i in f:
        if i[:2] == 'ال':
            i = i[2:]
        if i in keywords:
            keyword = i
    return (keyword,f)
```

```python
df = dataset.dropna()
dataset.poem_tags
```

```python
key = 'هجاء'
condition = df.poem_tags.str.startswith("قصائد "+key) | dataset.poem_tags.str.startswith("قصائد ال"+key)
poems = df[condition].reset_index(drop=True)
poems
```

```python
def get_poem(s):
    (key, f) = get_keyword(s)
    condition = df.poem_tags.str.startswith("قصائد " + key) | dataset.poem_tags.str.startswith("قصائد ال"+key)
    poems = df[condition].reset_index(drop=True)
    #let's check if the important info in the user input which is stored in f is ever mentioned in the poems_text
    #if so we want to take the one with the most frequency
    final = 0
    temp = 0
    poem = []
    #Note: it is not time efficient AT ALL 3 nested loops !! 
    #To Do: Find another method to search through them
    for p in poems.poem_text:
        temp = 0
        for w in p.split():
            for i in f:
                
                if i == w:
                    temp+=1
        if temp>=final:
            final = temp
            poem = p
    if final == 0:
        poem= poems.poem_text[random.randint(0,len(poems.poem_text)-1)]
    print(poem,'with matches to the prompt: ', final)
```

```python
s = input('أدخل ماذا تبغى كقصيدة:')
get_poem(s)
```
